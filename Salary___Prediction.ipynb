```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("salary.csv")

# IMPORTANT: Fix column names so code never breaks
dataset.columns = ['YearsExperience', 'Salary']

dataset.head()
```

```python
dataset.columns = ['YearsExperience', 'Salary']

dataset.head()
```

```python
X = dataset[['YearsExperience']]
y = dataset['Salary']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# sort so regression line becomes straight
sorted_index = X_test['YearsExperience'].argsort()
X_test_sorted = X_test.iloc[sorted_index]
y_pred_sorted = y_pred[sorted_index]

plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test_sorted, y_pred_sorted, color='red', linewidth=2, label='Predicted Line')

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary Prediction using Linear Regression")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python
exp = float(input("Enter Years of Experience: "))

# create dataframe with same column name as training
new_data = pd.DataFrame([[exp]], columns=['YearsExperience'])

predicted_salary = model.predict(new_data)

print("Predicted Salary =", predicted_salary[0])
```
